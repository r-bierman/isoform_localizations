```python
ensg_symbol = pd.read_csv('../processed_data/ENSG_symbol.txt')
ensg_symbol.head()
```

```python
matches = matches.drop_duplicates('hg38.kgXref.geneSymbol')
matches.head()
```

```python
matches = matches.merge(
    ensg_symbol,
    left_on='hg38.kgXref.geneSymbol',
    right_on='symbol',
    how='inner',
)

matches
```

```python
matches.drop(columns=['#hg38.knownGene.name','hg38.kgXref.geneSymbol'],inplace=True)
```

```python
matches.rename(columns={'hg38.kgXref.description':'description','hg38.kgXref.refseq':'refseq'},inplace=True)
```

```python
matches[['symbol','ENSG','description','refseq']]
```

```python
matches[['symbol','ENSG','description','refseq']].to_csv('../processed_data/gene_info.csv',index=False)
```
